```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
analytical = pd.read_csv('./new_result/analytical/qnet.plt', delimiter=' ')
cvode = pd.read_csv('./result_wrong_desktop/bepridil_0.0000_qnet_smp0.plt', delimiter=',')

plt.plot(analytical['Time'][:-1], analytical['Qnet'][:-1], label = 'Cluster')
plt.plot(cvode['Time'][:-1], cvode['Qnet'][:-1], label = 'Desktop')
plt.title("Qnet")

last_point = np.asarray(analytical['Qnet'])[-2], np.asarray(cvode['Qnet'])[-2]

plt.legend()

#analytical / cluster
plt.annotate(str(last_point[0]),(np.asarray(analytical['Time'])[-2], np.asarray(analytical['Qnet'])[-2]), 
             xytext=(np.asarray(analytical['Time'])[-2] - 200, np.asarray(analytical['Qnet'])[-2]-0.01),
             color='blue',
             fontsize=10
             )
#cvode / desktop
plt.annotate(str(last_point[1]),(np.asarray(cvode['Time'])[-2], np.asarray(cvode['Qnet'])[-2]),
             xytext=(np.asarray(cvode['Time'])[-2] - 200, np.asarray(cvode['Qnet'])[-2]-0.01),
             color='red',
             fontsize=10
             )


plt.show()
```
